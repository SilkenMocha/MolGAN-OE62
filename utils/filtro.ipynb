## Filtro 1

```python
import pickle
from rdkit import Chem
from rdkit.Chem.Draw import MolsToGridImage
import random
from IPython.display import display

# Ruta a tu archivo pickle (OE62.txt)
pickle_path = '/home/silkenmocha/Documentos/MolGAN-OE62/data/m1507656/OE62.txt'

# Carga el dict guardado (debe contener la clave 'smiles')
with open(pickle_path, 'rb') as f:
    data = pickle.load(f)

smiles_list = data.get('smiles', [])
print(f"Total de SMILES cargados: {len(smiles_list)}")
```

```python
# Celda 2: muestreo aleatorio
# Fija semilla si quieres reproducibilidad:
#random.seed(42)

n = 30  # número de moléculas a visualizar
sampled = random.sample(smiles_list, min(n, len(smiles_list)))
```

```python
# Celda 3: conversión y dibujo
# Convierte a Mol y filtra conversiones fallidas
mols = [Chem.MolFromSmiles(s) for s in sampled]
mols = [m for m in mols if m is not None]

# Dibuja en una cuadrícula 6×5 con subimágenes de 400×400 px
img = MolsToGridImage(mols, molsPerRow=6, subImgSize=(400, 400))

# Muestra en el notebook
display(img)
```

```python
# Convierte los objetos Mol de vuelta a SMILES
generated_smiles = [Chem.MolToSmiles(m) for m in mols]

# Mostrar los SMILES generados
for i, smi in enumerate(generated_smiles, 1):
    print(f"{i:2d}: {smi}")
```

```python
# Seleccionar la molécula en la posición 38 (índice 37)
idx = 22
mol_38 = mols[idx]

# Dibujarla individualmente usando MolsToGridImage con 1 en la fila
img_38 = MolsToGridImage([mol_38], molsPerRow=1, subImgSize=(400, 400))

# Mostrar en el notebook
display(img_38)
```
