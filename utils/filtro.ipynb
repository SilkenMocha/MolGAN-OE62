```python
from rdkit import Chem
from rdkit.Chem.Draw import MolsToGridImage
import random
from IPython.display import display

# Ruta a tu archivo .smi
smi_path = '/home/silkenmocha/Documentos/MolGAN-OE62/data/OE62_2.smi'

# Cargar SMILES desde el archivo .smi
with open(smi_path, 'r') as f:
    smiles_list = [line.strip() for line in f if line.strip()]

print(f"Total de SMILES cargados: {len(smiles_list)}")

# Convertir SMILES a molculas RDKit
molecules = [Chem.MolFromSmiles(s) for s in smiles_list if Chem.MolFromSmiles(s) is not None]

# Seleccionar molculas aleatorias para mostrar
random_mols = random.sample(molecules, min(25, len(molecules)))

# Mostrar en grid
img = MolsToGridImage(random_mols, molsPerRow=5, subImgSize=(300,300))
display(img)
```

### XYZ

```python
# %%
import json
import pandas as pd
from ase.io import read
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import MolsToGridImage
from io import StringIO
from ase.data import covalent_radii
import numpy as np

# %%
# Ruta al archivo JSON
path = '/home/silkenmocha/Documentos/MolGAN-OE62/data/m1507656/df_62k.json'

# Cargar el archivo JSON tipo tabla
with open(path, 'r') as f:
    data = json.load(f)

columnas = data["columns"]
filas = data["data"]
df = pd.DataFrame(filas, columns=columnas)

# Confirmar columnas
print("Columnas disponibles:", df.columns)

# %%
# Extraer la primera geometra XYZ no vaca
xyz_string = df['xyz_pbe_relaxed'].dropna().iloc[50]
print("Ejemplo de XYZ:\n", xyz_string)

# %%
# Convertir a objeto ASE
atoms = read(StringIO(xyz_string), format='xyz')

# Funcin para convertir objeto ASE a RDKit Mol
def ase_to_rdkit_mol(atoms_obj):
    symbols = atoms_obj.get_chemical_symbols()
    positions = atoms_obj.get_positions()

    mol_rw = Chem.RWMol()
    idx_map = []

    for sym in symbols:
        atom = Chem.Atom(sym)
        idx = mol_rw.AddAtom(atom)
        idx_map.append(idx)

    for i in range(len(symbols)):
        for j in range(i + 1, len(symbols)):
            d = np.linalg.norm(positions[i] - positions[j])
            r1 = covalent_radii[atoms_obj[i].number]
            r2 = covalent_radii[atoms_obj[j].number]
            if d < 1.2 * (r1 + r2):  # heurstica de enlace
                mol_rw.AddBond(i, j, Chem.BondType.SINGLE)

    mol = mol_rw.GetMol()

    conf = Chem.Conformer(len(symbols))
    for i, pos in enumerate(positions):
        conf.SetAtomPosition(i, list(pos))
    mol.RemoveAllConformers()
    mol.AddConformer(conf)

    return mol

# %%
# Convertir y visualizar
mol_rdkit = ase_to_rdkit_mol(atoms)
MolsToGridImage([mol_rdkit], molsPerRow=1, subImgSize=(300, 300))
```
